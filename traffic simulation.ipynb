```python
import random
import math
```

```python
class Car:
    def __init__(self,t,m,s,k):
        self.arrival=t
        self.number=m
        self.speed=s
        self.start_up_time=k
```

```python
class Cross:
    length=10
    def __init__(self,n,t):
        self.name=n
        self.pass_time=t
```

```python
class Road:
    def __init__(self,n,l):
        self.name=n
        self.length=l
```

```python
#at the very begining when the car entering the peachstreet
class Activity_EnterRoad:
    def __init__(self,car,road):
        self.car=car
        self.end_time=car.arrival
```

```python
class Activity_OnTheRoad:
    def __init__(self,activity,road):
        self.car=car
        self.start_time=activity.end_time
        self.end_time=self.start_time+road.length/activity.car.speed
        
```

```python
class Activity_EnterCross:
    def __init__(self,activity,cross):
        self.car=activity.car
        self.start_time=activity.end_time
```

```python
#class Activity_OnTheCross:
class Activity_OnTheCross:
    def __init__(self,activity_EnterCross,cross):
        self.car=activity_EnterCross.car
        self.start_time=activity_EnterCross.start_time
        self.end_time=self.start_time+cross_passing_time(self.car,cross)
```

```python
#generate random car arrival time
def generate_arrival_list(m):
    time_list=[0]
    for i in range(m-1):
        time_list.append(time_list[i]+random.randrange(1,60,1))
    return time_list
```

```python
#generate random average speed of a car
def generate_speed_list(m):
    speed_list=[]
    for i in range(m):
        speed_list.append(0.1*random.randrange(100,300,1))
    return speed_list
```

```python
#generate random start_up time of a car
def generate_start_time_list(m):
    start_list=[]
    for i in range(m):
        start_list.append(0.1*random.randrange(10,50,1))
    return start_list
```

```python
# assume the traffic light red=20s green=15s ,initial state is red
def light_is_green(t):
    t_red=20
    t_green=15
    t_total=t_red+t_green
    re=t%t_total
    if re>=0 and re<t_red:
        return True
    else:
        return False
 
```

```python
#compute the time for a Car to pass the Cross
def cross_passing_time(car,cross):
    return car.start_up_time+cross.length/car.speed
    
```
